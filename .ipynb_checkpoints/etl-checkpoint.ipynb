## Ethplorer API (https://etherscan.io/):

Endpoint: http://api.etherscan.io/

Documentation: https://github.com/EverexIO/Ethplorer/wiki/Ethplorer-API

```python
list_df = []

for wallet_id in wallet_ids.wallet_id:
    json_obj = get_eth_json_obj(wallet_id)
    temp_df = json_to_df(json_obj)
    list_df.append(temp_df)
    time.sleep(30)
```

```python
df_eth = pd.concat(list_df)
```

```python
df_eth.to_csv('data/df_etherscan.csv')
```

## OpenSea API:

```python
URL = 'https://api.opensea.io/api/v1/collections?asset_owner={}&offset=0&limit=300'
```

```python
list_df_opeansea = []

for wallet_id in wallet_ids.wallet_id:
    r = requests.get(URL.format(wallet_id))
    json_obj = r.json()
    temp_dfs = []
    for temp_df in json_obj:
        traits = temp_df.pop('traits')
        try:
            seller_fee = temp_df['fees'].pop('seller_fees')   
            seller_fee = list(seller_fee.values())[0]
        except:
            seller_fee = np.nan
        try:
            opensea_fee = temp_df['fees'].pop('opensea_fees') 
            opensea_fee = list(opensea_fee.values())[0]
        except:
            opensea_fee = np.nan
        temp_df = pd.json_normalize(temp_df)
        temp_df.loc[:, 'address'] = wallet_id
        temp_df.loc[:, 'traits'] = [traits]
        temp_df.loc[:, 'seller_fees'] = seller_fee
        temp_df.loc[:, 'opensea_fees'] = opensea_fee
        temp_dfs.append(temp_df)
    temp_df = pd.concat(temp_dfs)
    list_df_opeansea.append(temp_df)
    time.sleep(1)
```

```python
len(list_df_opeansea)
```

```python
df_opensea = pd.concat(list_df_opeansea)
```

```python
df_opensea.to_csv('data/df_opensea.csv')
```
